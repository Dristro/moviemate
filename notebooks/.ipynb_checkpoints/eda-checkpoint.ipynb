### Casting consistency

```python
top_cast = list(set(data_df.top_cast.str.split("|").explode()))
top_cast[:3]
```

```python
import re

def normalize(tok: str) -> str:
    """
    1. lowers
    2. remove dots and stuff
    3. keep letters and spaces only

    Args:
        tok (str): token/sequence to normalize
    Returns:
        tok (str), token normalized
    """
    tok = tok.lower().strip()
    tok = re.sub(r"\.", "", tok)
    tok = re.sub(r"[^a-z\s]", "", tok)
    tok = re.sub(r"\s+", " ", tok)
    return tok

top_cast = (
    data_df
        .top_cast
        .str
        .split("|")
        .explode()
        .map(normalize)
        .unique()
        .tolist()
)
top_cast[:5]
```

```python
from rapidfuzz import process, fuzz

def find_variants(target, choices, cutoff = 85):
    target = normalize(target)
    matches = process.extract(
        target,
        choices,
        scorer=fuzz.token_set_ratio,
        score_cutoff=cutoff,
        limit=None,
    )
    return [m[0] for m in matches]

variants = find_variants("John Oliver", top_cast)
variants
```

```python
def same_person(a, b):
    a_parts = a.split()
    b_parts = b.split()
    if not a_parts or not b_parts:
        return False

    # match last name
    if a_parts[-1] != b_parts[-1]:
        return False

    # first name, exact or initial
    return (
        a_parts[0] == b_parts[0] or
        a_parts[0].startswith(b_parts[0]) or
        b_parts[0].startswith(a_parts[0])
    )

def find_variants_strict(target, choices):
    target_n = normalize(target)
    return [
        n for n in choices
        if same_person(target_n, n)
    ]
```

```python
from tqdm.auto import tqdm
from rapidfuzz import fuzz

clusters = []
used = set()

for name in tqdm(top_cast):
    if name in used:
        continue

    group = []
    for other in top_cast:
        if other in used:
            continue

        if (
            fuzz.token_set_ratio(name, other) > 85 and
            same_person(name, other)
        ):
            group.append(other)
            used.add(other)

    clusters.append(group)
```

```python
# See if there exists a cluster with more than one element
any([len(clusters[i]) > 1 for i in range(len(clusters))])
```

```python
cluster_ids = [i for i in range(len(clusters)) if len(clusters[i]) > 1]
cluster_ids
```

```python
[clusters[i] for i in cluster_ids]
```

```python
# mapping built using chatgpt given the set of countries and languages from ds
country_lang = {
    "us": {"english", "spanish"},
    "gb": {"english"},
    "au": {"english"},
    "ca": {"english", "french"},
    "nz": {"english", "maori"},
    "ie": {"english", "irish"},
    
    "in": {"hindi", "tamil", "telugu", "malayalam", "kannada", "bengali", "marathi", "punjabi", "english"},
    "bd": {"bengali"},
    "np": {"nepali"},
    "lk": {"sinhalese", "tamil"},
    
    "cn": {"mandarin", "cantonese"},
    "hk": {"cantonese", "mandarin", "english"},
    "sg": {"mandarin", "english", "malay", "tamil"},
    "jp": {"japanese"},
    "kr": {"korean"},
    "th": {"thai"},
    "id": {"indonesian"},
    "kh": {"khmer"},
    "vn": {"vietnamese"},
    
    "fr": {"french"},
    "be": {"dutch", "french", "german"},
    "ch": {"german", "french", "italian"},
    "de": {"german"},
    "at": {"german"},
    "nl": {"dutch"},
    "se": {"swedish"},
    "dk": {"danish"},
    "fi": {"finnish", "swedish"},
    "no": {"norwegian"},
    "is": {"icelandic"},
    
    "es": {"spanish", "catalan", "galician"},
    "pt": {"portuguese"},
    "it": {"italian"},
    "pl": {"polish"},
    "cz": {"czech"},
    "sk": {"slovak"},
    "hu": {"hungarian"},
    "gr": {"greek"},
    "ru": {"russian"},
    "ua": {"ukrainian"},
    
    "br": {"portuguese"},
    "mx": {"spanish"},
    "ar": {"spanish"},
    "ve": {"spanish"},
    
    "za": {"afrikaans", "zulu", "xhosa", "english"},
    "bw": {"tswana"},
    "mw": {"chichewa; nyanja"},
    
    "ae": {"arabic"},
    "kw": {"arabic"},
    "lb": {"arabic", "french"},
    "ma": {"arabic"},
    "ly": {"arabic"},
    
    "tr": {"turkish"},
}
```

```python
def normalize(x):
    if isinstance(x, str):
        return x.lower().strip()
    return x

def rule_outlier(row):
    c = row["origin_country"]
    l = row["spoken_languages"]

    if pd.isna(c) or not isinstance(l, list):
        return False

    expected = country_lang.get(c)
    if not expected:
        return False

    if any(lang.strip() in expected for lang in l):
        return False
    return True


working_df = data_df.copy()
working_df["spoken_languages"] = (
    working_df["spoken_languages"]
    .str.lower()
    .str.split("|")
)

working_df["origin_country"] = working_df["origin_country"].apply(normalize)
working_df["spoken_languages"] = working_df["spoken_languages"].apply(normalize)
working_df["rule_outlier"] = working_df[["origin_country", "spoken_languages"]].apply(rule_outlier, axis=1)

print((working_df["rule_outlier"]).value_counts())

flagged_ids = working_df.rule_outlier
working_df[flagged_ids][["origin_country", "spoken_languages"]]
```

## Token statistics

We will use nltk to analyse token distribution.
Since most LLMs use some flavor of BPE, for a
model with a large enough vocab-size, its safe
to assume most common words are tokens. So we
will tokenize all text into word tokens and
count their statistics. This will help us find
the best chunk-size.

```python
avg_tagline = sum(data_df.tagline.fillna("").str.len())/data_df.shape[0]
avg_overview = sum(data_df.overview.str.len())/data_df.shape[0]
print(f"Average number of chars in `tagline`: {avg_tagline:.2f}")
print(f"Average number of chars in `overview`: {avg_overview:.2f}")
print(f"Average number of chars in `tagline + overview`: {(avg_tagline + avg_overview):.2f}")
```

```python
import nltk
nltk.download('punkt_tab')
```

I plan on using the following columns to create each chunk.
* `overview`
* `tagline`
* `top_cast`
* `director`
* `title`

All other columns/attrs will be fetched using the code generation
model that will write a query to the DB to fetch relevant inforamtion
at runtime.

```python
import string
from nltk.tokenize import word_tokenize

sequences = []
for overview in tqdm(data_df[["title", "overview", "tagline", "top_cast", "director"]].fillna("").agg(" ".join, axis=1)):
    tokens = word_tokenize(overview.lower())
    sequences.append([t for t in tokens if t not in string.punctuation])

seq_lens = [len(seq) for seq in sequences]
seq_lens[:5]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=seq_lens, kde=True)
plt.axvline(sum(seq_lens)/len(seq_lens), linestyle="--")

plt.xlabel("Token length")
plt.ylabel("Frequency")
plt.text(
    sum(seq_lens)/len(seq_lens),
    plt.ylim()[1] * 0.8,
    f"Mean: {sum(seq_lens)/len(seq_lens):.2f}",
    rotation=-45,
)
```

```python
import numpy as np

idx = np.argmax(seq_lens)
print("="*15 + " Movie with longest overview " + "="*15)
print(f"Max seq len: {max(seq_lens)}")
print(data_df.loc[idx])

del idx
```

So on **average**, most movie's infomration can be expressed in **65 tokens**.
The **max** tokens required to express a movie is **189**. Given that most
embedding models come with max seq-len support of **$512 \ge$**, so there
would be **no need to chunk** the data.
